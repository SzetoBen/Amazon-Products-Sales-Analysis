# Dataset Information:
- Size: The dataset has 42,675 product entries (rows) and 17 features (columns)
- Features: 
1) product_title (text)
2) product_rating (numerical)
3) total_reviews (numerical)
4) purchased_last_month (numerical)
5) discounted_price (numerical)
6) original_price (numerical)
7) is_best_seller (categorical)
8) is_sponsored (categorical)
9) has_coupon (categorical)
10) buy_box_availability (categorical)
11) delivery_date (temporal)
12) sustainability_tags (categorical)
13) product_image_url (text)
14) product_page_url (text)
15) data_collected_at (temporal)
16) product_category (categorical)
17) discount_percentage (numerical)

Total:
Numerical features: 6
Text features: 3
Categorical features:6
Temporal features: 2
- There are no duplicates within the dataset
- The following features having missing values: product_rating, total_reviews, purchased_last_month, discounted_price, original_price, buy_box_availability, delivery_date, sustainability_tags, product_page_url, discount_percentage 

- Distributions: We discovered that prices are highly skewed features with most products below $100 but a few outliers are priced at over $1000. Ratings are also heavily centered around 4-4.5 stars and not evenly distributed. The number of reviews for most products is under 50 while some outliers have over thousands of reviews.
- All min/max values are generally reasonable except for some prices being at $0 and some extreme discounts from 90-100% off prices. These values may be due to missing or listing data erros.
- To handle missing data, we will use median imputation for numerical features such as rating and discounts. For the categorical feautres, missing entries will be filled with "unknown". Rows that are missing multiple key features will be dropped.

# EDA Summary

Through EDA we were able to reveal that there is little correlation between the many numerical variables and that there is significant variability in product pricing and review counts features. We have also discovered that there are distinct patterns across product categories such as electronics and fashion have the most competitive pricing. We discovered that the data is relatively clean but requires preprocessing for missing and outlier handling.
